```python
import matplotlib.pyplot as plt
# draws charts smoothly
%matplotlib inline 
```

```python
x = [1,2,3,4,5,6,7,8]
y = [11,33,22,64,95,16,47,88]
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y,color='red', linewidth=3, linestyle='dashdot')
```

```python
# to label x and y axis
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Temperature Data')
plt.plot(x,y, linestyle='dotted')
```

```python
# format string
plt.plot(x,y, 'bo--') # color + shape
```

```python
# same thibg but with py args
# alpha -> controls visibility from 0 to 1, 0 being transparent and 1 being opaque
plt.plot(x,y, color='green', marker='D', linestyle='dashed', alpha=0.5)
```

```python
# Axes labels, Legend, Grid

days = [1,2,3,4,5,6,7]
max_t = [50, 51,53,54,55,58,60]
min_t = [30,31,33,34,36,39,32]
avg_t = [40, 41, 43, 44, 45, 48, 46]
```

```python
# plotting all in one chart
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(days, max_t, label='Max')
plt.plot(days, min_t, label='Min')
plt.plot(days, avg_t, label='Avg')

plt.legend()

# to show grid on the canvas
plt.grid()
```
