```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blood_sugar = [113,85,90, 100, 60, 150, 93, 120, 125, 70, 79, 92, 135]
```

```python
plt.hist(blood_sugar, bins=[80, 100,125,200], rwidth=0.8, histtype='step')
```

```python
# more than one static data
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]
plt.ylabel('Number of Pateints')
plt.xlabel('Sugar Level')
plt.title('Blood Sugar Analysis')
plt.hist([blood_sugar_men,blood_sugar_women], bins=[80, 100,125,200], rwidth=0.8, color=['b','r'], label=['Men', 'Women'])

plt.legend()
```

```python
# plotting histograms horizontally
plt.ylabel('Number of Pateints')
plt.xlabel('Sugar Level')
plt.title('Blood Sugar Analysis')
plt.hist([blood_sugar_men,blood_sugar_women], bins=[80, 100,125,200], rwidth=0.95, color=['b','r'], label=['Men', 'Women'], orientation='horizontal')

plt.legend()
```
