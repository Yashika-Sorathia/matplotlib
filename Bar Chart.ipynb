```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
company = ['GOOGLE', 'AMZN', 'APPLE', 'FB']
revenue = [90, 120, 150, 30]
profit = [20, 30, 60, 10]
```

```python
# since bar chart do not take stirng value create an numpy array

xpos = np.arange(len(company))
xpos
```

```python
# convert ypos back to string
plt.xticks(xpos, company)
plt.ylabel('Revenue(in billion)')
plt.bar(xpos, revenue)
```

```python
# ploting more bars into same canvas

# since x axis is same for both revenue and profit, data will be shown on same bar
plt.xticks(xpos, company)
plt.bar(xpos-0.2, revenue, label='Revenue', width=0.4)
plt.bar(xpos+0.2, profit, label='Profit', width=0.4)

plt.legend()
```

```python
# Horizontal bar chart
plt.yticks(xpos, company)
plt.barh(xpos-0.2, revenue, label='Revenue', height=0.4)
plt.barh(xpos+0.2, profit, label='Profit',height=0.4)

plt.legend()
```
