```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
```

```python
plt.pie(exp_vals, labels=exp_labels, radius=1.5, autopct='%0.2f%%', shadow=True, explode=[0,0.3,0,0,0])
# to save chart as file
# always call plt.savefig() before plt.show() since it creates new canvase, which results into blank image
plt.savefig('pie_chart.png', bbox_inches='tight', pad_inches=0.5, transparent=True)
plt.show()
```
